```python
import pandas as pd
```

```python
# Pulling in the players stats dataset
plyrs = pd.read_csv('players.csv')
plyrs.head()
```

```python
plyrs = plyrs.set_index('_id')
```

```python
plyrs.head()
```

```python
# Pulling in the players salaries dataset
salary = pd.read_csv('salaries_1985to2018.csv')
salary.head()
```

```python
salary.info()
```

```python
# Dropping the unnecessary columns in player salaries
salary = salary.drop(['league', 'season', 'season_end', 'season_start', 'team'], axis=1)
salary.head()
```

```python
# Combining the salaries into a career earnings, reducing the number of duplicate enteries.
agg_functions = {'player_id':'first', 'salary':'sum'}
career_earn = salary.groupby(salary['player_id']).aggregate(agg_functions)
```

```python
career_earn.set_index('player_id')
```

```python
comb_plyrs_stats = pd.concat([plyrs, career_earn], axis=1)
```

```python
comb_plyrs_stats.info()
```

```python
comb_plyrs_stats = comb_plyrs_stats.drop(['birthDate',
                                          'birthPlace',
                                          'college',
                                          'draft_pick',
                                          'career_PER',
                                          'career_TRB',
                                          'career_WS',
                                          'career_eFG%',
                                          'draft_round',
                                          'draft_team',
                                          'draft_year',
                                          'position',
                                          'height', 
                                          'highSchool', 
                                          'name', 
                                          'shoots', 
                                          'weight', 
                                          'player_id'], axis=1)
```

```python
clean_plyrs_stats = comb_plyrs_stats.dropna()
```

```python
clean_plyrs_stats.info()
```

```python
remove_char = clean_plyrs_stats[clean_plyrs_stats['career_FG%'] == '-'].index
clean_plyrs_stats.drop(remove_char, inplace = True)
```

```python
remove_char1 = clean_plyrs_stats[clean_plyrs_stats['career_FG3%'] == '-'].index
clean_plyrs_stats.drop(remove_char1, inplace = True)
```

```python
remove_char2 = clean_plyrs_stats[clean_plyrs_stats['career_FT%'] == '-'].index
clean_plyrs_stats.drop(remove_char2, inplace = True)
```

```python
clean_plyrs_stats.head()
```

```python
clean_plyrs_stats['career_FG%'] = clean_plyrs_stats['career_FG%'].astype(float)
clean_plyrs_stats['career_FG3%'] = clean_plyrs_stats['career_FG3%'].astype(float)
clean_plyrs_stats['career_FT%'] = clean_plyrs_stats['career_FT%'].astype(float)
```

```python
plyers_csv_stats = clean_plyrs_stats.to_csv('prepared_plyrs_stats.csv', index = True)
print('\nCSV String:\n', plyers_csv_stats)
```
